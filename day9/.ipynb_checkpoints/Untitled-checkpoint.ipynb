```python
import sys
from math import cos, sqrt, pi #DCT 구현을 위한 코사인, root, pi
import numpy as np #binary 데이터열을 처리하기 위해서
import matplotlib.pyplot as plt #이미지를 파이썬 코드 안에서 보기 위해서
```

```python
def dct_type2(x, y, u, v, n):
    def alpha(a):
        if a==0:
            return sqrt(1.0/n)
        else:
            return sqrt(2.0/n)
        return alpha(u) * alpha(v) * cos(((2*x+1)*(u*pi))/(2*n)) * cos(((2*y+1)*(v*pi))/(2*n))
    #DCT Type-2 로직
    
def process_dct(height, width, image):
    processed_image = image
    for y in range(0, height):
        for x in range(0, width):
            processed_image[y, x] = dct_type2(x, y, height, width, 256)
    return processed_image
```

```python
with open('lena.raw', 'rb') as f: #raw 이미지를 rb (바이너리로 읽기 모드)로 연다
    height = 256 #가로 픽셀
    width = 256 #세로 픽셀 -> 하드코딩이 아니라 meta데이터로써 받을 방법이 있을지 고민해보기
    bin_image = np.fromstring(f.read(), dtype=np.uint8) # np 배열로 만들어서 처리하기
    bin_image.shape = (height, width)
    plt.imshow(bin_image, cmap='gray') #color map은 gray임.
    plt.show(block=True) #이미지가 보임.
    after_image = process_dct(256, 256, bin_image)
    plt.imshow(after_image, cmap='gray') #color map은 gray임.
    plt.show(block=True) #이미지가 보임.
```
